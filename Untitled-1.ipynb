```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the datasets using the exact filenames from your sidebar
sentiment_df = pd.read_csv('fear_greed_index.csv')
trader_df = pd.read_csv('historical_data.csv')

# 2. Part A, Task 1: Document rows/columns (Requirement: Part A)
print("--- Data Dimensions ---")
print(f"Sentiment Data: {sentiment_df.shape[0]} rows, {sentiment_df.shape[1]} columns")
print(f"Trader Data: {trader_df.shape[0]} rows, {trader_df.shape[1]} columns")

# 3. Check for missing values (Requirement: Part A)
print("\n--- Missing Values ---")
print("Sentiment missing values:\n", sentiment_df.isnull().sum())
print("\nTrader missing values:\n", trader_df.isnull().sum())
```

```python
# 1. Clean column names to remove hidden spaces
trader_df.columns = trader_df.columns.str.strip()
sentiment_df.columns = sentiment_df.columns.str.strip()

# 2. Part A, Task 2: Convert timestamps and align datasets by date
# Handling the 'Timestamp IST' from your file and 'date' from sentiment
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], errors='coerce').dt.date
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'], errors='coerce').dt.date

# Remove any rows that failed the date conversion
trader_df = trader_df.dropna(subset=['Timestamp IST'])
sentiment_df = sentiment_df.dropna(subset=['date'])

# 3. Align datasets (Merge)
merged_df = pd.merge(trader_df, sentiment_df[['date', 'classification']], 
                     left_on='Timestamp IST', right_on='date', how='left')

# 4. Part A, Task 3: Ensure metrics are numeric for analysis
merged_df['Closed PnL'] = pd.to_numeric(merged_df['Closed PnL'], errors='coerce')
merged_df['Size USD'] = pd.to_numeric(merged_df['Size USD'], errors='coerce')

print("--- Alignment and Data Prep Complete ---")
print(merged_df[['Timestamp IST', 'Account', 'classification']].head())
```

```python
# 1. Part B, Q1: Does performance (PnL) differ between Fear vs Greed?
performance = merged_df.groupby('classification')['Closed PnL'].mean()
print("--- Avg PnL by Sentiment (Evidence) ---")
print(performance)

# 2. Part B, Q2: Do traders change behavior (Trade Size) based on sentiment?
behavior_size = merged_df.groupby('classification')['Size USD'].mean()
print("\n--- Avg Trade Size by Sentiment ---")
print(behavior_size)

# 3. Part B, Q2: Trade Frequency (Number of trades per day)
frequency = merged_df.groupby('classification').size()
print("\n--- Trade Frequency by Sentiment ---")
print(frequency)
```

```python
# 1. Part B, Task 3: Identify 2-3 segments (Large vs Small traders)
median_size = merged_df['Size USD'].median()
merged_df['Segment'] = merged_df['Size USD'].apply(lambda x: 'Large' if x > median_size else 'Small')

# 2. Part B, Task 4: Visualization (Insights backed by charts)
plt.figure(figsize=(10,5))
performance.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average PnL per Market Sentiment')
plt.ylabel('Profit (USD)')
plt.xticks(rotation=45)
plt.show()

# Segment performance chart
segment_analysis = merged_df.groupby(['Segment', 'classification'])['Closed PnL'].mean().unstack()
segment_analysis.plot(kind='bar', figsize=(12,6))
plt.title('Performance Comparison: Large vs Small Traders')
plt.ylabel('Average PnL')
plt.show()
```

```python
# Save final merged results for submission
merged_df.to_csv('final_trader_analysis_report.csv', index=False)

# Summary for Part C (Rules of Thumb):
# 1. During Extreme Greed, Large Traders should maintain aggressive positions (Avg PnL ~486).
# 2. During Extreme Fear, Small Traders should reduce frequency to avoid low-profit volatility.

print("Project Analysis Complete. File 'final_trader_analysis_report.csv' is ready.")
```
